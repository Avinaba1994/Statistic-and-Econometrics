```python
import random
dice = [1,2,3,4,5,6]
z=[]
def MC_Dice(n):
    for i in range(n):
        a= random.choice(dice)
        b= random.choice(dice)
        c= a+b
        z.append(c)
    return z
MC_Dice(1000)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.hist(z)
plt.show()
```
